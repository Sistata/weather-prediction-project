```python
from sklearn.linear_model import Ridge

rr = Ridge(alpha=.1)
```

```python
predictors = weather.columns[~weather.columns.isin(["target", "name", "station"])]
```

```python
def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions = []
    
    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]
        
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(weather, rr, predictors)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mean_absolute_error(predictions["actual"], predictions["prediction"])
```

```python
# Our model shows, on average, we were 4 degree off from the correct temperature. So, in order to improve the accuracy of our model, we can take the average of past few days and looking how the current day compared to those days i.e adding more predictors 

def pct_diff(old, new):
    return(new - old) / old

def compute_rolling(weather, days, col):
    label = f'rolling_{days}_{col}'
    
    weather[label] = weather[col].rolling(days).mean()
    weather[f"{label}_pct"] = pct_diff(weather[label], weather[col])
    return weather

rolling_days = [3, 14]

for days in rolling_days:
    for col in ["tmax", "tmin", "prcp"]:
        weather = compute_rolling(weather, days, col)
```
